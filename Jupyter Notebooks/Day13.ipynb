```python
# Importing necessary libraries for analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Getting the Dataset
milkshake_ratings = pd.read_csv(r'D:\Data Journey\Python-Summer-Party\DataSets\milkshake_ratings.csv')

# Display the first few rows to understand the data
print(milkshake_ratings.head(10))  # Shows the first 5 rows with columns
print(list(milkshake_ratings.columns))
print('Number of rows and columns is:', milkshake_ratings.shape)
```

```python
# Question One
# Cleaning to remove any duplicate entries in the customer ratings data 
cleanded_milkshake_ratings = milkshake_ratings.drop_duplicates()
cleanded_milkshake_ratings
```

```python
# Question Two : average customer rating for each milk flavor
# 0. Start from original DataFrame and drop duplicates
milkshake_ratings_cleaned = milkshake_ratings.drop_duplicates().copy()

# 1. Clean the 'rating' column: remove spaces, replace commas with periods, convert to numeric
milkshake_ratings_cleaned['rating'] = pd.to_numeric(milkshake_ratings_cleaned['rating'], errors='coerce')

# Optional: drop rows where 'rating' is NaN before averaging
milkshake_ratings_cleaned = milkshake_ratings_cleaned.dropna(subset=['rating'])

# 2. Calculate average rating per milkshake flavor
avg_rating_per_flavor = (milkshake_ratings_cleaned.groupby('flavor')['rating'].mean().reset_index(name='avg_rating'))

# 3. Merge the average ratings back into the cleaned DataFrame
milkshake_ratings_with_avg = milkshake_ratings_cleaned.merge(avg_rating_per_flavor, on='flavor', how='left')

milkshake_ratings_with_avg
```

```python
# Question Three
# difference between that customer's rating and the average rating for the flavor.
milkshake_ratings_with_avg['rating_diff'] = milkshake_ratings_with_avg['rating'] - milkshake_ratings_with_avg['avg_rating']
milkshake_ratings_with_avg
```
